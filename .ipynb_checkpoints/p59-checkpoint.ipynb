```python
import pandas as pd
```

```python
# read file
with open('p059_cipher.txt') as file:
    message = file.read().split(',')
```

Key has length 3 so let's divide the encrypted message into 3 blocks
- block 0 contains the letters that appears in position 0, 3, 6, 9, ...
- block 1 contains the letters that appears in position 1, 4, 7, 10, ...
- block 2 contains the letters that appears in position 2, 5, 8, 11, ...

```python
blocks = [[], [], []]
for i in range(len(message)):
    blocks[i%3].append(int(message[i]))
```

```python
lower_case = range(97, 123)
upper_case = range(65, 91)
```

```python
def can_be_english(a):
    res = {}
    n = len(a)    
    
    # create frequency dictionary
    freq_dic = {}
    for ch in a:
        if ch in freq_dic:
            freq_dic[ch] += 1
        else:
            freq_dic[ch] = 1
    
    # check frequency of characters
    e = ['e', 'E']
    count = 0
    for ch in e:
        try:
            count += freq_dic[ch]
        except:
            count += 0
            
    res['e'] = count/n
    
    try:
        res['space'] = freq_dic[' ']/n
    except:
        res['space'] = 0
    
    
    # e,t,a,o,i,n should compose 40% of the dictionary
    freq_ch = ['e', 't', 'a', 'o', 'i', 'n', 'E', 'T', 'A', 'O', 'I', 'N']
    count = 0
    for ch in freq_ch:
        try:
            count += freq_dic[ch]
        except KeyError:
            count += 0
            
    res['e,t,a,o,i,n'] = count/n

    # letters should compose 80% of the dictionary
    count = 0
    for ch in a:
        if (ord(ch)>=65 and ord(ch)<91) or (ord(ch)>=97 and ord(ch)<123):
            count +=1
    
    res['letters'] = count/n
    
    return res
```

## Block 0

```python
possible_dec = {}
for key in lower_case:
    possible_dec[key] = [chr(en^key) for en in blocks[0]]    
```

```python
data = []
for p in possible_dec:
    row = [p]
    res = can_be_english(possible_dec[p])
    for key in res:
        row.append(res[key])
    data.append(row)
```

```python
df = pd.DataFrame(data, columns=['key', 'e', 'space', 'e,t,a,o,i,n', 'letters'])
df.sort_values(by=['e', 'space', 'e,t,a,o,i,n', 'letters'], ascending=False)
```

101 could be the key!

```python
dec_blocks = [[], [], []]
```

```python
for i in range(len(blocks[0])):
    dec_blocks[0].append(chr(int(blocks[0][i])^101))
```

```python
dec_m = ''
idx = 0
for i in range(len(message)):
    if i%3==0:
        dec_m += dec_blocks[0][idx]
        idx+=1
    else:
        dec_m += '*'
```

## Block 2

```python
possible_dec = {}
for key in lower_case:
    possible_dec[key] = [chr(en^key) for en in blocks[2]]    
```

```python
data = []
for p in possible_dec:
    row = [p]
    res = can_be_english(possible_dec[p])
    for key in res:
        row.append(res[key])
    data.append(row)
```

112 could be the key!

```python
for i in range(len(blocks[2])):
    dec_blocks[2].append(chr(int(blocks[2][i])^112))
```

```python
dec_m = ''
idx_0 = 0
idx_1 = 0
idx_2 = 0
for i in range(len(message)):
    if i%3==0:
        dec_m += dec_blocks[0][idx_0]
        idx_0+=1
    elif i%3==1:
        dec_m += dec_blocks[1][idx_1]
        idx_1+=1   
    elif i%3==2:
        dec_m += dec_blocks[2][idx_2]
        idx_2+=1  
```

```python
s = 0
for block in dec_blocks:
    for ch in block:
        s += ord(ch)
```
